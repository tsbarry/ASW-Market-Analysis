## Standard Salaries Visualizations & statistics

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# read csv file
clean_df= pd.read_csv('data/clean_data.csv')
clean_df.shape

# Histogram of salary_standardized
sns.histplot(clean_df['salary_standardized'], kde= True) 
plt.savefig('images/salary_standardized.jpg')
```

```python
# Boxplot of salary_standardized
sns.boxplot(clean_df['salary_standardized'])
plt.savefig('images/sal_stand_boxplot.jpg')
```

```python
# Kolmogorov_Smirnov Test
stats.kstest(clean_df['salary_standardized'],stats.norm.cdf, args=(clean_df['salary_standardized'].mean(),clean_df['salary_standardized'].std()))
```

## Remote and Non_remote Standard Salaries Visualizations & Statistics

```python
# Remote jobs Histogram
remote_df = clean_df[clean_df['work_from_home']== True]
sns.histplot(data= remote_df, x = 'salary_standardized', kde= True) 
```

```python
# Kolmogorov_Smirnov Test for remote salaries
stats.kstest(remote_df['salary_standardized'],stats.norm.cdf, args=(remote_df['salary_standardized'].mean(),remote_df['salary_standardized'].std()))
```

```python
# Non_remote jobs Histogram
non_remote_df = clean_df[clean_df['work_from_home']== False]
sns.histplot(data= non_remote_df, x= 'salary_standardized', kde= True)
```

```python
#Kolmogorov_Smirnov Test for remote salaries
stats.kstest(non_remote_df['salary_standardized'],stats.norm.cdf, args=(non_remote_df['salary_standardized'].mean(),non_remote_df['salary_standardized'].std()))
```

```python
# side by side histogram for remote v non_remote
sns.histplot(data= remote_df, x = 'salary_standardized', kde= True) 
sns.histplot(data= non_remote_df, x= 'salary_standardized', kde= True)
plt.legend(loc= 'upper right', labels = ['remote', 'non_remote'])
plt.savefig('images/remote_and_non_remote.jpg') 
```

## Comparison of Remote v. Non_remote

```python
# Boxplot of remote v non_remote work
sns.boxplot(data= clean_df, y = 'salary_standardized', x='work_from_home')
plt.savefig('images/work_from_home_boxplot.jpg')
```

```python
# KDE Plot of remote v non_remote work
sns.kdeplot(data= clean_df, x= 'salary_standardized', hue = 'work_from_home')
plt.savefig('images/kde_work_from_home.jpg')
```

```python
# T-test
remote_df = clean_df[clean_df['work_from_home']== True]
non_remote_df = clean_df[clean_df['work_from_home']== False]
stats.ttest_ind(remote_df['salary_standardized'], non_remote_df['salary_standardized'])
```
