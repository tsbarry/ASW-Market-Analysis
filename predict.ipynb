```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 

# Read csv file
clean_df= pd.read_csv('data/clean_data.csv')

# create a linear regression model
regr = LinearRegression()
X = clean_df.loc[:,'airflow':'word']
regr.fit(X, clean_df['salary_standardized'])
predict= regr.predict(X)

# R2 value
regr.score(X, clean_df['salary_standardized'])

plot= sns.histplot(data=predict, kde= True, bins = 30)
plot.set(xlim=(20_000, 250000))
plot.set(ylim=(0, 1200))
plt.title('Linear regression of Salary Standardized')
plt.savefig('images/Linear_Regression.jpg')
```
